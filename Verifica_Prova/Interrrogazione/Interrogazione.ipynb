```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("netflix1.csv", encoding = "utf8")
righe = file.readlines()
file.close()
```

```python
righe
```

```python
dati = {"Title":[], "IMDb_score":[], "Metacritic_score":[], "Awards_Received":[], "Awards_Nominated":[], "IMDb_votes":[] }
```

```python
for riga in righe[1:]:
    lista_Campi = riga[:-1].split(",")
    dati["Title"].append(lista_Campi[0])
    dati["IMDb_score"].append(float(lista_Campi[1]))
    dati["Metacritic_score"].append(float(lista_Campi[2]))
    dati["Awards_Received"].append(float(lista_Campi[3]))
    dati["Awards_Nominated"].append(float(lista_Campi[4]))
    dati["IMDb_votes"].append(float(lista_Campi[5]))
    
for nome, d in dati.items():
    dati[nome] = np.array(d)
dati
```

```python
colori = np.random.uniform(0.,1.,(10,3))
```

```python
fig, axs = plt.subplots(nrows=1 , ncols=2 , figsize = (7,4))
axs[0].scatter(dati["IMDb_score"], dati["Awards_Received"], marker = "v")
axs[0].set_xlabel("IMDb_score", fontsize = 10)
axs[0].set_ylabel("Awards_Received", fontsize = 10)
axs[0].set_title("Confronto 1", fontsize = 15)

axs[1].scatter(dati["IMDb_score"], dati["IMDb_votes"], marker = "2")
axs[1].set_xlabel("IMDb_score", fontsize = 10)
axs[1].set_ylabel("Awards_votes", fontsize = 10)
axs[1].set_title("Confronto 2", fontsize = 15)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (7,7))
ax.scatter(dati["IMDb_score"], dati["Awards_Received"], marker = "v", s = dati["IMDb_votes"]/1000, c = "orange")
ax.set_xlabel("IMDb_score", fontsize = 10)
ax.set_ylabel("Awards_Received", fontsize = 10)
ax.set_title("Confronto", fontsize = 15)
plt.show
```
