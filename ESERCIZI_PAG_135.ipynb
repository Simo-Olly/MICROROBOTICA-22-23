```python
numero_secondi = 7200
x = [0]
y = [0]
movimenti = np.random.choice(["est", "ovest", "nord", "sud"], numero_secondi)
print(movimenti)
j = 0
for i in movimenti:
    if i == "est":
        x.append(x[j] + 1)
        y.append(y[j])
    elif i == "ovest":
        x.append(x[j] - 1)
        y.append(y[j])
    elif i == "nord":
        x.append(x[j])
        y.append(y[j] + 1)
    elif i == "sud":
        x.append(x[j])
        y.append(y[j] - 1)
    j += 1   
```

```python
fig, ax = plt.subplots(figsize = (6, 6))
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize = (6, 6))
ax.scatter(x, y, s = 1.0)
```

# ES 11

```python
dati = {"velv":[], "baro":[], "terma":[], "igro":[], "pulv":[]}
headers = ["velv", "baro", "terma", "igro", "pulv"]
posizioniCampi = [5, 8, 9, 10 , 13]

with open("stazioneMetereologica.csv", "r") as f:
    righe = f.readlines()
    for riga in righe:
        valori = riga[:-1].split(",")
        valoriUtili = [float(valori[i]) for i in posizioniCampi]
        if 9999.9 not in valoriUtili: 
            for p, h in zip(posizioniCampi, headers):
                dati[h].append(float(valori[p]))
```

```python

ncols = 5
nrows = 5
fig, axs = plt.subplots(ncols = ncols, nrows = nrows, figsize = (16, 16), constrained_layout = True)
for j, dato_y in enumerate(headers):
    for i, dato_x in enumerate(headers):
        if i != j:
            axs[i, j].scatter(dati[dato_x], dati[dato_y], c = "y", s = 10)
            axs[j, i].set_xlabel(dato_x, fontsize = 14)
            axs[j, i].set_ylabel(dato_y, fontsize = 14)
        else:
            axs[j, i].hist(dati[dato_x], color = "y", linewidth=0.5, edgecolor="black")
            axs[j, i].set_xlabel(dato_x, fontsize = 14)
            
fig.suptitle('Misure metereologiche', fontsize = 25)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.hist(dati["pulv"], color = "orange", edgecolor = "black", bins = 60)
ax.set_xlabel("pulv", fontsize = 14)
ax.set_ylabel("n° giorni", fontsize = 14)
ax.set_title("Distribuzione Piogge", fontsize = 20)
plt.show()
```
