### ES stili

```python
x = np.linspace(0., 4*np.pi, 100)
s1 = np.sin(x)
s2 = np.cos(x)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, s1, linestyle = ":", linewidth = 5, color = "red", label = "sin")
ax.plot(x, s2, linestyle = "-.", linewidth = 5, color = "green", label = "cos")
ax.set_xlabel("x", fontsize=20)
ax.set_ylabel("y", fontsize=20)
ax.set_title("Titolo", fontsize = 20)
ax.tick_params(axis="both", labelsize=10)
ax.legend()
ax.grid()
plt.savefig("./grafico.pdf")
plt.show()
```

```python
file = open("C:\\Users\\Marco\\Desktop\\5Â° superiore\\MICROROBOTICA\\Es_Python\\Cartella_fileCSV\\penguins_size.csv", "r")
righe = file.readlines()
file.close()

dati = {"species": [], "island" : [], "culmen_length_mm" : [], "culmen_depth_mm" : [], "flipper_length_mm" : [], "body_mass_g" : [], "sex" : [] }

for riga in righe[1:]:
    lista_Campi = riga[:-1].split(",")
    
    if "NA" not in lista_Campi:
        dati["species"].append(lista_Campi[0])
        dati["island"].append(lista_Campi[1])
        dati["culmen_length_mm"].append(float(lista_Campi[2]))
        dati["culmen_depth_mm"].append(float(lista_Campi[3]))
        dati["flipper_length_mm"].append(float(lista_Campi[4]))
        dati["body_mass_g"].append(float(lista_Campi[5]))
        dati["sex"].append(lista_Campi[6])
```

```python
for nome_colonna, lista in dati.items():
    dati[nome_colonna] = np.array(lista)
```

```python
dati
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.hist(dati["body_mass_g"])
plt.show()
```

## Esercizio punti nel cerchio

```python
n_punti = 10000
array_X = np.random.uniform(-1,1,n_punti)
array_Y = np.random.uniform(-1,1,n_punti)
```

```python
fig, ax = plt.subplots(figsize = (8,8))
ax.scatter(array_X, array_Y, s = 1)
plt.show()
```

```python
ipotenusa = np.sqrt(array_X**2+ array_Y**2)
cerchio_x = array_X[ipotenusa <= 1]
cerchio_y = array_Y[ipotenusa <= 1]
colori = np.zeros(n_punti)
colori[ipotenusa <= 1] = 1
```

```python
fig, ax = plt.subplots(figsize = (8,8))
ax.scatter(array_X, array_Y, s = 1)
ax.scatter(cerchio_x, cerchio_y, s = 1, c = "red")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,8))
ax.scatter(array_X, array_Y, s = 1, c = colori, cmap = "gnuplot")
plt.show()
```

```python
Nc = np.count_nonzero(ipotenusa < 1)
Nc
```
