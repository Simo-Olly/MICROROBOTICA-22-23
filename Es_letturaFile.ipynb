```python
dizionario = {}
dizionario_1 = {}
caratteri_non_voluti = ["", ",", ".", "(", ")", ";", "?", "!"]

for r in righe:
    r = r[:-1]
    r = r.lower()
    r.replace("", " ")
    lista = [lettera for lettera in r if lettera not in caratteri_non_voluti]
    riga_pulita = "".join(lista)
    lista = riga_pulita.split(" ")
    for i in lista:
        if i not in dizionario:
            dizionario[i] = 1
        else:
            dizionario[i] += 1
    print(lista)
```
