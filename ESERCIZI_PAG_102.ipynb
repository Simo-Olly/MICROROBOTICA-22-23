```python
import uuid
import numpy as np
```

```python
n_passeggeri = 200000
temperature = np.random.normal(36.5, 0.25, n_passeggeri)
identificativi =  np.array([str(uuid.uuid4()) for _ in range(n_passeggeri)])
print(temperature[:5]) # visualizza le prime 5 temerature generate 
print(identificativi[:5]) # visualizza i primi 5 uuid generati
```

```python
%%timeit
contatore = 0
for t in temperature:
    if t > 37.5:
        contatore += 1
print(f"Il numero di asseggeri con temperatura maggiore di 37.5 °C. è {contatore}\n")

print("ATTENZIONE i passeggeri con temperatura maggiore di 37.5 °C:")
for i in range(n_passeggeri):
    if temperature[i] > 37.5:
        print(identificativi[i])
```

```python
%%timeit

print("Il numero di passeggeri con con temperatura maggiore di 37.5 °C. è:", np.sum(temperature > 37.5))
print()

print("ATTENZIONE i passeggeri con temperatura maggiore di 37.5 °C:")
print(identificativi[temperature > 37.5])
```
