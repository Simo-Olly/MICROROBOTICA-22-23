```python
n = 1
l = [1]
while n < 100:
    i = 2
    p = False
    while i <= n:
        if n % i == 0 and p == False:
            l.append(n)
            p = True
        i += 1
    n += 1  
```

```python
n=0
i=0
l = []
while n < 100:
    i += 1
    count = 1
    for j in range(2, i):
        if i % j == 0:
            count = 0
            break
    if count == 1:
        l.append(i)
        n += 1
np.array(l)
```
